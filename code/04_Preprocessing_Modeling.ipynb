```python
# import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
import os

from PIL import Image
```

### Get Pixel Data From Images

```python
def create_list_files(dir):
    r = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            r.append(os.path.join(root, name))
    return r

list_files = create_list_files('../images/styles/')
list_files
```

```python
partial_list_files = list_files[:10]
```

```python
def get_img_num_from_filename(input_str):
    temp_num_str=""
    for character in str(input_str[-11:-4]):        #ignore last 4 chars because it's always '.jpg'
        if character in "0123456789":
            temp_num_str += character
    return int(temp_num_str)

# get_img_num_from_filename('../images/styles/Abstract-Art\\image_48689.jpg')
```

```python
# Reference: https://www.hackerearth.com/practice/notes/extracting-pixel-values-of-an-image-in-python/

def get_pixel_data_once(file_name):
    im = Image.open(file_name,  'r')
    pix_val = list(im.getdata())
    print(f"pix_val={pix_val}")
    pix_val_flat = [x for sets in pix_val for x in sets]
    return pix_val_flat

# get_pixel_data_once('../images/styles/Abstract-Art\\image_48689.jpg')

def get_pixel_data_batch(list_files):
    df_of_pixel_data = pd.DataFrame()

    # for file in list_files:
    for file_name in list_files:
        # print(file_name)
        img_num = get_img_num_from_filename(file_name)
        pix_val_flat = get_pixel_data_once(file_name)
        pix_val_flat_plus_img_num = pix_val_flat + [img_num]
        
        df_of_pixel_data = pd.concat([df_of_pixel_data, pd.DataFrame(pix_val_flat_plus_img_num).T])
    
    df_of_pixel_data.rename(columns={df_of_pixel_data.shape[1]-1:'img_num'}, inplace=True)    #we appended img_num as last col. rename last col
    df_of_pixel_data.set_index('img_num', inplace=True)

    return df_of_pixel_data
```

```python
get_pixel_data_once('../images/styles/Abstract-Art\\image_48689.jpg')
```

```python
df_of_pixel_data = get_pixel_data_batch(partial_list_files)
df_of_pixel_data
```

```python
df_of_pixel_data.to_csv('../data/df_of_pixel_data.csv', index = True)
```
